```python
import matplotlib.pyplot as plt
import json

filename = 'td_minimax_4_pure'
```

```python
with open (filename + '.outcome.json') as f:
    data = json.load(f)
```

```python
len(data)
```

```python
xs = []
ys = []
for i in range(0, len(data), 100):
    pts = data[i:i+100]
    mean = sum(pts) / len(pts)
    xs.append(i)
    ys.append(mean)
```

```python
fig = plt.figure(figsize=(15,4))
plt.title('TD Agent v. Minimax Agent, Utility per 100 Episodes', fontsize=18)
plt.plot(xs, ys, '.-', color='#980000', lw=1, label='a')
# plt.legend()
```

```python
fig.savefig(filename + '_util.png', dpi=200)
```

```python
wins = []
loses = []
for i in range(0, len(data), 100):
    pts = data[i:i+100]
    w = 0
    l = 0
    for pt in pts:
        if pt > 0:
            w += 1
        if pt < 0:
            l += 1
    w /= len(pts) / 100
    l /= len(pts) / 100
    wins.append(w)
    loses.append(l)
```

```python
fig = plt.figure(figsize=(15,4))
plt.title('TD Agent v. Minimax Agent, Win/Loss per 100 Episodes', fontsize=18)
plt.plot(xs, wins, '.-', lw=1, label='wins', color='#980000')
plt.plot(xs, loses, '.-', lw=1, label='losses', color='#003262')
plt.legend()
```

```python
fig.savefig(filename + '_winloss.png', dpi=200)
```
